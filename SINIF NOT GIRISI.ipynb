```python
def mat1ort():
    toplam = 0
    for i in listematx:
        toplam = toplam + i
    return toplam/len(listematx)  
def tur1ort():
    toplam = 0
    for i in listeturx:
        toplam = toplam + i
    return toplam/len(listeturx)

listeisim = []
listemat1 = []
listetur1 = []
listematx = []
listeturx = []
listeorta = []
kont = True
while kont:
    isim = input("Öğrenci adını giriniz ")
    if (isim == "stop" or isim == "STOP" or isim == "."):
        kont = False
    else:
        uzunluk = 25-len(isim)
        isim = isim + uzunluk*" "
        listeisim.append(isim)
        matnot1 = input("Matematik 1. sınav notunu giriniz ")
        matnotx = int(matnot1)
        listematx.append(matnotx)
        if (matnot1 != "100"):
            matnot1 = " " + matnot1
        listemat1.append(matnot1)
        turnot1 = input("Türkçe 1. sınav notunu giriniz ")
        turnotx = int(turnot1)
        listeturx.append(turnotx)
        if (turnot1 != "100"):
            turnot1 = " " + turnot1
        listetur1.append(turnot1)
        dersort = (matnotx + turnotx)/2
        listeorta.append(dersort)
        
sayı = len(listeisim)       
sınort = (mat1ort() + tur1ort())/2
index = 0
print(100*"*")
print("ÖĞRENCİ İSMİ",12*" ","MATEMATİK 1. YAZILI",10*" ","TÜRKÇE 1. YAZILI",10*" ","DERS ORTALAMASI")
print(100*"*")
while (index < len(listeisim)):
    print("{}            {}                      {}                      {}".format(listeisim[index],listemat1[index],listetur1[index],listeorta[index]))
    index = index + 1

print(sayı,"ÖĞRENCİNİN",10*" ","MATEMATİK ORTALAMASI",10*" ","TÜRKÇE ORTALAMASI",10*" ","SINIF ORTALAMASI")
print("                                    {}                     {}                     {}".format(round(mat1ort(),2),round(tur1ort(),2),round(sınort,2)))
```
